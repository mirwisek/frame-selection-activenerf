```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Graph Results

```python
def plot_psnr_lpips_ssim(logs_path, subject, k):
    psnr_vals_dict = {}
    lpips_vals_dict = {}
    ssim_vals_dict = {}

    # Read the data from the npy file
    psnr_data1 = np.load(f"{logs_path}/results/psnr_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    psnr_data2 = np.load(f"{logs_path}/results/psnr_vals_dict_clustering.npy", allow_pickle=True).item()
    psnr_data3 = np.load(f"{logs_path}/results/psnr_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data1 = np.load(f"{logs_path}/results/lpips_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data2 = np.load(f"{logs_path}/results/lpips_vals_dict_clustering.npy", allow_pickle=True).item()
    lpips_data3 = np.load(f"{logs_path}/results/lpips_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data1 = np.load(f"{logs_path}/results/ssim_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data2 = np.load(f"{logs_path}/results/ssim_vals_dict_clustering.npy", allow_pickle=True).item()
    ssim_data3 = np.load(f"{logs_path}/results/ssim_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()

    for strategy, values in psnr_data1.items():
        psnr_vals_dict[strategy] = values
    
    for strategy, values in psnr_data3.items():
        psnr_vals_dict["iou"] = values

    for strategy, values in psnr_data2.items():
        psnr_vals_dict["embedding"] = values
    

    for strategy, values in lpips_data1.items():
        lpips_vals_dict[strategy] = values
    
    for strategy, values in lpips_data3.items():
        lpips_vals_dict["iou"] = values

    for strategy, values in lpips_data2.items():
        lpips_vals_dict["embedding"] = values
    
    

    for strategy, values in ssim_data1.items():
        ssim_vals_dict[strategy] = values

    for strategy, values in ssim_data3.items():
        ssim_vals_dict["iou"] = values
    
    for strategy, values in ssim_data2.items():
        ssim_vals_dict["embedding"] = values
    
    
    
    # Plot PSNR, SSIM, and LPIPS in subplots
    plt.figure(figsize=(15, 4))

    plt.subplot(131)
    for strategy, values in psnr_vals_dict.items():
        plt.plot(values, label="PSNR for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("PSNR for " + subject + " k=" + str(k))

    plt.subplot(132)
    for strategy, values in lpips_vals_dict.items():
        plt.plot(values, label="LPIPS for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("LPIPS for " + subject + " k=" + str(k))

    plt.subplot(133)
    for strategy, values in ssim_vals_dict.items():
        plt.plot(values, label="SSIM for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("SSIM for " + subject + " k=" + str(k))

    plt.tight_layout()
    plt.savefig(f"results/psnr_lpips_ssim_{subject}_k{k}.png")
```

```python
input_budgets = [5, 10, 15, 20, 25]

subjects = ['chair', 'drums', 'ficus', 'hotdog', 'lego', 'materials', 'mic', 'ship']

for k in input_budgets:
        for key in subjects:
            logs_path = f"logs/{key}_k{k}"
            (plot_psnr_lpips_ssim(logs_path, key, k))
```

### Average Graphs

```python
def get_average_psnr_lpips_ssim(logs_path, subject, k):

    psnr_vals_dict = {}
    lpips_vals_dict = {}
    ssim_vals_dict = {}

    # Read the data from the npy file
    psnr_data1 = np.load(f"{logs_path}/results/psnr_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    psnr_data2 = np.load(f"{logs_path}/results/psnr_vals_dict_clustering.npy", allow_pickle=True).item()
    psnr_data3 = np.load(f"{logs_path}/results/psnr_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data1 = np.load(f"{logs_path}/results/lpips_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data2 = np.load(f"{logs_path}/results/lpips_vals_dict_clustering.npy", allow_pickle=True).item()
    lpips_data3 = np.load(f"{logs_path}/results/lpips_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data1 = np.load(f"{logs_path}/results/ssim_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data2 = np.load(f"{logs_path}/results/ssim_vals_dict_clustering.npy", allow_pickle=True).item()
    ssim_data3 = np.load(f"{logs_path}/results/ssim_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()

    for strategy, values in psnr_data1.items():
        psnr_vals_dict[strategy] = values
    
    for strategy, values in psnr_data3.items():
        psnr_vals_dict["iou"] = values

    for strategy, values in psnr_data2.items():
        psnr_vals_dict["embedding"] = values
    

    for strategy, values in lpips_data1.items():
        lpips_vals_dict[strategy] = values
    
    for strategy, values in lpips_data3.items():
        lpips_vals_dict["iou"] = values

    for strategy, values in lpips_data2.items():
        lpips_vals_dict["embedding"] = values
    
    

    for strategy, values in ssim_data1.items():
        ssim_vals_dict[strategy] = values

    for strategy, values in ssim_data3.items():
        ssim_vals_dict["iou"] = values
    
    for strategy, values in ssim_data2.items():
        ssim_vals_dict["embedding"] = values

    return psnr_vals_dict, lpips_vals_dict, ssim_vals_dict
```

```python
input_budgets = [5, 10, 15, 20, 25]

subjects = ['chair', 'drums', 'ficus', 'hotdog', 'lego', 'materials', 'mic', 'ship']

for k in input_budgets:
    global_psnrs = {}
    global_lpips = {}
    global_ssims = {}

    for key in subjects:
        psnrs_dict, lpips_dict, ssims_dict = get_average_psnr_lpips_ssim(f"logs/{key}_k{k}", key, k)

        for strategy, values in psnrs_dict.items():
            if strategy not in global_psnrs:
                global_psnrs[strategy] = []
            global_psnrs[strategy].append(values)
        
        for strategy, values in lpips_dict.items():
            if strategy not in global_lpips:
                global_lpips[strategy] = []
            global_lpips[strategy].append(values)
        
        for strategy, values in ssims_dict.items():
            if strategy not in global_ssims:
                global_ssims[strategy] = []
            global_ssims[strategy].append(values)

    average_psnrs = {}
    average_lpips = {}
    average_ssims = {}

    for strategy, values in global_psnrs.items():
        average_psnrs[strategy] = np.mean(np.array(values), axis=0)   

    for strategy, values in global_lpips.items():
        average_lpips[strategy] = np.mean(np.array(values), axis=0)

    for strategy, values in global_ssims.items():
        average_ssims[strategy] = np.mean(np.array(values), axis=0)

    # Plot PSNR, SSIM, and LPIPS in subplots
    plt.figure(figsize=(15, 4))

    plt.subplot(131)
    for strategy, values in average_psnrs.items():
        plt.plot(values, label="PSNR for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("PSNR for " + " k=" + str(k))

    plt.subplot(132)
    for strategy, values in average_lpips.items():
        plt.plot(values, label="LPIPS for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("LPIPS for " + " k=" + str(k))

    plt.subplot(133)
    for strategy, values in average_ssims.items():
        plt.plot(values, label="SSIM for " + strategy + ": {:.4f}".format(values[-1]))
    plt.legend()
    plt.title("SSIM for " + " k=" + str(k))

    plt.tight_layout()
    plt.savefig(f"results/average_psnr_lpips_ssim_k{k}.png")
```

```python
def get_psnr_lpips_ssim(logs_path, subject, k):
    psnr_vals_dict = {}
    lpips_vals_dict = {}
    ssim_vals_dict = {}

    # Read the data from the npy file
    psnr_data1 = np.load(f"{logs_path}/results/psnr_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    psnr_data2 = np.load(f"{logs_path}/results/psnr_vals_dict_clustering.npy", allow_pickle=True).item()
    psnr_data3 = np.load(f"{logs_path}/results/psnr_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data1 = np.load(f"{logs_path}/results/lpips_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    lpips_data2 = np.load(f"{logs_path}/results/lpips_vals_dict_clustering.npy", allow_pickle=True).item()
    lpips_data3 = np.load(f"{logs_path}/results/lpips_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data1 = np.load(f"{logs_path}/results/ssim_vals_dict_baseline_{subject}_k{k}.npy", allow_pickle=True).item()
    ssim_data2 = np.load(f"{logs_path}/results/ssim_vals_dict_clustering.npy", allow_pickle=True).item()
    ssim_data3 = np.load(f"{logs_path}/results/ssim_vals_dict_iou_{subject}_k{k}.npy", allow_pickle=True).item()

    for strategy, values in psnr_data1.items():
        psnr_vals_dict[strategy] = values

    for strategy, values in psnr_data3.items():
        psnr_vals_dict["iou"] = values

    for strategy, values in psnr_data2.items():
        psnr_vals_dict["embedding"] = values

    for strategy, values in lpips_data1.items():
        lpips_vals_dict[strategy] = values

    for strategy, values in lpips_data3.items():
        lpips_vals_dict["iou"] = values
    
    for strategy, values in lpips_data2.items():
        lpips_vals_dict["embedding"] = values

    for strategy, values in ssim_data1.items():
        ssim_vals_dict[strategy] = values

    for strategy, values in ssim_data3.items():
        ssim_vals_dict["iou"] = values

    for strategy, values in ssim_data2.items():
        ssim_vals_dict["embedding"] = values

    psnr_max_values = {}
    lpips_max_values = {}
    ssim_max_values = {}

    for strategy, values in psnr_vals_dict.items():
        psnr_max_values[strategy] = round(values[-1],2)
    for strategy, values in lpips_vals_dict.items():
        lpips_max_values[strategy] = round(values[-1],2)
    for strategy, values in ssim_vals_dict.items():
        ssim_max_values[strategy] = round(values[-1],2)

    # give me a df with max

    return psnr_max_values, lpips_max_values, ssim_max_values
```

```python
def get_csv(data, method, k):

    # Initialize an empty dictionary to store the data
    data_dict = {}

    # Initialize an empty list to store the current values
    current_values = []

    # For each item in the data list
    for item in data:
        # If the item is a string, it's a subject
        if isinstance(item, str):
            # If there are any current values, add them to the data dictionary under the previous subject
            if current_values:
                data_dict[previous_subject] = current_values
                current_values = []
            # Update the previous subject
            previous_subject = item
        # If the item is a number, it's a value
        else:
            current_values.append(item)

    # Don't forget to add the last set of values to the data dictionary
    if current_values:
        data_dict[previous_subject] = current_values

    methods = ['camera','random','fvs_distance', 'fvs_iou','embedding']

    df = pd.DataFrame(data_dict)
    df.index = methods

    # print(df)
    df.to_csv(f'results/results_{method}_k{k}.csv')
    # return df
```

```python
input_budgets = [5, 10, 15, 20, 25]

subjects = ['chair', 'drums', 'ficus', 'hotdog', 'lego', 'materials', 'mic', 'ship']

for k in input_budgets:
        p_values = []
        l_values = []
        s_values = []

        for key in subjects:
            p_values.append(key)
            l_values.append(key)
            s_values.append(key)

            logs_path = f"logs/{key}_k{k}"
            psnr_vals_dict, lpips_vals_dict, ssim_vals_dict = get_psnr_lpips_ssim(logs_path, key, k)
            for method, vals in psnr_vals_dict.items():
                p_values.append(vals)
            for method, vals in lpips_vals_dict.items():
                l_values.append(vals)
            for method, vals in ssim_vals_dict.items():
                s_values.append(vals)
            
        get_csv(p_values,'psnr',k)
        get_csv(l_values,'lpips',k)
        get_csv(s_values,'ssim',k)
```
